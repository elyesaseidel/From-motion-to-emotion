```python
import aspect_based_sentiment_analysis as absa
```

```python
nlp = absa.load() #instaniate the pipeline
text_1 = ("this is all wrong I shouldn't be up here I should be back in school on the other side of the ocean yet you all come to us young people for Hope how dare you you have stolen my dreams and my childhood with your empty words and yet I'm one of the lucky ones people are suffering people are dying entire ecosystems are collapsing we are in the beginning of a mass extinction and all you can talk about is the money and fairy tales of Eternal economic growth how dare you")

people, dreams, childhood, ecosystem, mass_extinction, fairy_tales, economic_growth  = nlp(text_1, aspects=['people', 'dreams', 'childhood', 'ecosystem', 'mass_extinction', 'fairy_tales', 'economic_growth'])
```

```python
print("people", people.sentiment)
print("dreams", dreams.sentiment)
print("childhood", childhood.sentiment)
print("ecosystem", ecosystem.sentiment)
print("mass exstinction", mass_extinction.sentiment)
print("fairy tales", fairy_tales.sentiment)
print("ecnonomic growth", economic_growth.sentiment)
```

```python
nlp = absa.load() #instaniate the pipeline
text_2 = ("thank you thank you thank you thank you to the academy for this all 6,000 members thank you to the other nominees all these performances were impeccable in my opinion I didn't see a false note anywhere I want to thank valet or director Jennifer garnered with daily there's a few things about three things to my account that I need each day one of them is something to look up to another is something to look forward to in another is someone to chase now first off I want to thank God because that's who I look up to he's great my life with opportunities that I know are not of my hand or any other human hand he is showing me that it's a scientific fact that gratitude reciprocates")

academy, members, nominees, performances, god, scientific_fact, gratitude = nlp(text_2, aspects=['academy', 'members', 'nominees', 'performances', 'god', 'scientific_fact', 'gratitude']) # manually choose the aspects based on the domain
```

```python
# sentiment according to chosen aspects
print("academy", academy.sentiment)
print("members", members.sentiment)
print("nominees", nominees.sentiment)
print("performances", performances.sentiment)
print("god", god.sentiment)
print("scientific fact", scientific_fact.sentiment)
print("gratitude", gratitude.sentiment)
```

```python
nlp = absa.load() #instaniate the pipeline
text_3 = ("I used to think the whole purpose of life was pursuing happiness everyone said the path to happiness was success so I searched for that ideal job that perfect boyfriend that beautiful apartment but instead of ever feeling fulfilled I felt anxious and the drift and I wasn't alone my friends they struggled with this too")

happiness, success, job, boyfriend, apartment, feeling, friends  = nlp(text_3, aspects=['happiness', 'success', 'job', 'boyfriend', 'apartment', 'feeling', 'friends'])
```

```python
print("happiness", happiness.sentiment)
print("success", success.sentiment)
print("job", job.sentiment)
print("boyfriend", boyfriend.sentiment)
print("apartment", apartment.sentiment)
print("feeling", feeling.sentiment)
print("friends", friends.sentiment)
```

```python
import aspect_based_sentiment_analysis as absa

recognizer = absa.aux_models.BasicPatternRecognizer()
nlp = absa.load(pattern_recognizer=recognizer)
```

```python
# importance 
completed_task = nlp(text_3, aspects=['happiness', 'success', 'job', 'boyfriend', 'apartment', 'feeling', 'friends'])
happiness, success, job, boyfriend, apartment, feeling, friends = completed_task.examples

absa.summary(happiness)
absa.display(happiness.review)

absa.summary(success)
absa.display(success.review)

absa.summary(job)
absa.display(job.review)

absa.summary(boyfriend)
absa.display(boyfriend.review)

absa.summary(apartment)
absa.display(feeling.review)

absa.summary(friends)
absa.display(friends.review)
```
